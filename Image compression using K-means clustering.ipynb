```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = plt.imread('saad.jpg')
plt.imshow(img)
plt.axis('off')
plt.show
```

```python
type(img)
```

```python
print(img.shape)
print(img.size)
```

```python
w,h,d =img.shape
image_array = img.reshape(w*h,d)
print(image_array.shape)
image_array = image_array/255
```

```python
from sklearn.utils import shuffle
image_array_sasmple = shuffle(image_array,random_state = 1)[:1000]
```

```python
image_array_sasmple
```

```python
kmeans = KMeans(n_clusters=6, random_state=1)
kmeans.fit(image_array_sasmple)
```

```python
labels = kmeans.predict(image_array)
labels
```

```python
print(kmeans.cluster_centers_)
c=kmeans.cluster_centers_
```

```python
# Create Orignal image Accourding to labels 
```

```python
def recreate_image(c,labels,w,h,d):
    image=np.zeros((w,h,d))
    label_idx=0
    for i in range(w):
        for j in range(h):
            image[i][j] = c[labels[label_idx]]
            label_idx+=1
    return(image)
```

```python
plt.figure(1)
plt.axis('off')
plt.title('orignal')
plt.imshow(img)
plt.show()
plt.figure(2)
plt.axis('off')
plt.title('Reducve')
plt.imshow(recreate_image(c,labels,w,h,d))
plt.show()
```
